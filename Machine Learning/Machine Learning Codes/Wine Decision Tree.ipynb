```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
wine
```

```python
df=pd.DataFrame(wine["data"],columns=['alcohol',
  'malic_acid',
  'ash',
  'alcalinity_of_ash',
  'magnesium',
  'total_phenols',
  'flavanoids',
  'nonflavanoid_phenols',
  'proanthocyanins',
  'color_intensity',
  'hue',
  'od280/od315_of_diluted_wines',
  'proline'])
```

```python
df
```

```python
dft=pd.DataFrame(wine["target"],columns=["Target"])
```

```python
dft
```

```python
dfwine=pd.concat([df,dft],axis=1)
```

```python
dfwine
```

```python
X=dfwine.drop("Target",axis=1)
```

```python
y=dfwine["Target"]
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kf=KFold(n_splits=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc_score=cross_val_score(dtc,X,y,cv=kf)
```

```python
dtc_score.mean()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(X)
```

```python
dtc1=DecisionTreeClassifier(criterion="entropy",min_samples_split=10,max_depth=5)
```

```python
ss_X=ss.transform(X)
```

```python
dtc1_score=cross_val_score(dtc1,ss_X,y,cv=kf)
```

```python
dtc1_score.mean()
```

```python
from sklearn.svm import SVC
```

```python
svc_score=cross_val_score(SVC(),ss_X,y,cv=kf)
```

```python
svc_score.mean()
```
