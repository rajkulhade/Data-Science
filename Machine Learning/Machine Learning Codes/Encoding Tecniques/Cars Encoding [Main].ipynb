## Data Encoding

First we will import python Libraries like numpy and pandas for  Data Encoding 
firstly what is Data Encoding?

we can say like we are converting Categorical Data
(which are not in Numeric Form like words,strings etc) into Numerical Data.

```python
import pandas as pd
import numpy as np
```

```python
Cars=pd.read_csv(r"C:\Users\admin\Desktop\DATA SCIENCE\Machine Learning\machine learning dataset\One Hot Encoding\cars.csv")
```

```python
Cars.columns
```

#### In this dataset you can see km_driven and selling_price so we dont need to convert them

so lets convert the columns["brand"] using the get dummies

```python
pd.get_dummies(Cars["brand"])
```

but you can see here it has created many columns as i have said you in word file this is ordinal Data so we should use Ordinal Encoding in this
and it is independent so Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
```

Never forget that Ordinal Encoder uses only 2D data so with your dataframe and columns use double Columns like i have used Cars[["brand"]]

```python
oe.fit(Cars[["brand"]])
```

```python
ary1=oe.transform(Cars[["brand"]])
```

```python
ary1 # its the 2D array lets convert is to DataFrame
```

```python
encode=pd.DataFrame(ary1,columns=["New_Brand"]) # give the column name as same column give as new so you will understnd which one to drop
```

```python
encode
```

Lets concat now

```python
df=pd.concat([Cars,encode],axis=1)
```

```python
df
```

Lets drop now the brand column which is Categorical which is no longer needed now

```python
df.drop("brand",axis=1,inplace=True)
```

```python
df
```

#### Now you have seen the Ordinal Encoding now for owner column you can Get_dummies as well as OneHotEncoder()

lets do the OneHotEncoder as you have seen i have did it before . In other dataset i will Use it Dont worry

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
ohe.fit(df[["owner"]])
```

```python
ary2=ohe.transform(df[["owner"]]).toarray()
```

```python
ary2
```

```python
encode2=pd.DataFrame(ary2,columns=["New_Owner"])
```

why this eror is coming guys you now bcoz i was giving only onw column name but i told you in get dummies and OneHotEncoder you will get more columns 

```python
ohe.categories_
```

see how it has numbered out categorical data as first Owner as 0,Fourth And above Owner as 2 and so on 

```python
encode2=pd.DataFrame(ary2,columns=['First Owner', 'Fourth & Above Owner', 'Second Owner',
        'Test Drive Car', 'Third Owner'])
```

```python
encode2
```

```python
df=pd.concat([df,encode2],axis=1)
```

```python
df
```

```python
df.drop("owner",axis=1,inplace=True)
```
